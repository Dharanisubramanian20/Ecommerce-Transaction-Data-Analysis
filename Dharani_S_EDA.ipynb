## customer sign-up by region

```python
region_counts=customers['Region'].value_counts()
region_counts
```

```python
import matplotlib.pyplot as plt
region_counts.plot(kind="bar")
plt.title("customer sign-up by region")
plt.xlabel("Region")
plt.ylabel("Number of sign-ups")
plt.show()
```

## sign-up date over time

```python
customers['SignupDate']=pd.to_datetime(customers['SignupDate'])
signups_over_time=customers.set_index('SignupDate').resample('M').size()
signups_over_time.plot()
plt.title('customer signups over time')
plt.xlabel('Month')
plt.ylabel('No of sign-ups')
plt.show()
```

## Product prices

```python
products['Price'].hist(bins=30)
plt.title('Product Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Products by category


```python
category_counts = products['Category'].value_counts()
category_counts.plot(kind='bar')
plt.title('Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()
```

## Transaction amounts

```python
import matplotlib.pyplot as plt
import seaborn as sns

transactions['TotalValue'].hist(bins=30)
plt.title('Transaction Value Distribution')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.show()
```

## Transactions over time

```python
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
transactions.set_index('TransactionDate').resample('M').size().plot()
plt.title('Monthly Transactions')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.show()
```
